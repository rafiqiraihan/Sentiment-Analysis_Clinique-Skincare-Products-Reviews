```python
import pandas as pd
import os
```

```python
reviews_files = sorted([f for f in os.listdir() if f.startswith("reviews_") and f.endswith(".csv")])

# menggabungkan semua file reviews
all_reviews = pd.concat(
    [pd.read_csv(os.path.join(file)) for file in reviews_files],
    ignore_index=True
)

print(f"Total reviews: {len(all_reviews):,}")
all_reviews.head()
```

```python
all_reviews.info()
```

```python
products = pd.read_csv("product_info.csv")
```

```python
products.info()
```

```python
product_cols_to_add = [
    'product_id',
    'brand_id',
    'loves_count',
    'size',
    'variation_type',
    'variation_value',
    'variation_desc',
    'ingredients',
    'value_price_usd',
    'sale_price_usd',
    'limited_edition',
    'new',
    'online_only',
    'out_of_stock',
    'sephora_exclusive',
    'highlights',
    'primary_category',
    'secondary_category',
    'tertiary_category',
    'child_count',
    'child_max_price',
    'child_min_price'
]

products_clean = products[product_cols_to_add]
```

```python
merged_df = pd.merge(all_reviews, products_clean, on="product_id", how="left")
```

```python
merged_df.info()
```

```python
merged_df.to_csv("reviews_with_product_info.csv", index=False)
```

```python
merged_df['brand_name'].value_counts().head()
```

```python
# Memilih salah satu brand untuk di analisis
select_brand = "CLINIQUE"

clinique_df = merged_df[merged_df['brand_name'] == select_brand].copy()
clinique_df.head()
```

```python
# Menyimpan dataset brand
clinique_df.to_csv(f"reviews_{select_brand.replace(' ', '_').lower()}.csv", index=False)
```
